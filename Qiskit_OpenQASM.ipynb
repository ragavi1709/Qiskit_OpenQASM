```python
# Qiskit 2.x Example â€“ OpenQASM Demonstration

from qiskit import QuantumCircuit, transpile
from qiskit_aer import AerSimulator
from qiskit.qasm3 import dumps as qasm3_dumps
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
qc = QuantumCircuit(3, 3)
qc.h(0)
qc.h(1)
qc.h(2)

# Apply CNOT gates to create entanglement
qc.cx(0, 1)
qc.cx(1, 2)

# Apply a phase gate to one qubit (just for variation)
qc.p(0.5, 2)

# Step 3: Add a barrier
qc.barrier()

# Step 4: Add measurement to classical bits
qc.measure([0, 1, 2], [0, 1, 2])
```

```python
print("\nğŸ§© Modified 3-Qubit Circuit:")
print(qc.draw('text'))
```

```python
qasm_code = qasm3_dumps(qc)
print("\nğŸ“œ Generated OpenQASM 3 Code:\n")
print(qasm_code)
```

```python
simulator = AerSimulator()
compiled = transpile(qc, simulator)
job = simulator.run(compiled, shots=1024)
result = job.result()
counts = result.get_counts()

# Show results
print("\nğŸ“Š Simulation Result (Counts):")
print(counts)

plot_histogram(counts, title="2-Qubit Entanglement Output")
plt.show()

# QASM syntax notes
print("\nğŸ” QASM 3 Syntax Notes:")
print("â€¢ Hadamard gate â†’ h q[0];")
print("â€¢ CNOT gate   â†’ cx q[0], q[1];")
print("â€¢ Barrier     â†’ barrier q[0], q[1];")
print("â€¢ Measurement â†’ measure q[0] -> c[0];")
```
